````python
def get_ord_arr_delay(df, airport):
    '''
    Takes a Pandas DataFrame and an airport
    and returns a Pandas Series of arrival delays.
    
    Parameters
    ----------
    df: A Pandas DataFrame with columns
        'Year', 'Month', 'Origin', 'DayofMonth', 'CRSDepTime', and 'ArrDelay'.
    airport: A str of three-letter airport code.
    
    Returns
    -------
    A pandas.Series object.
    '''
    
    # your code goes here
    df = df[df["Origin"] == airport]
    df.index = pd.to_datetime(df["Year"].astype(str) + "-" + df["Month"].astype(str) + "-" + df["DayofMonth"].astype(str) 
                          + " " + df["CRSDepTime"].astype(str), format = "%Y-%m-%d %H%M")
    result = df["ArrDelay"]
    result.index.name = "Date" # add a name to index 
    result = result.dropna(axis = 0, how = "any") # remove rows with NaN
    # my code ends here
    
    return result

arr_delay = get_ord_arr_delay(df, 'ORD')
````
