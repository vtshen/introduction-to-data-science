````python
def pivot_table(df):
    '''
    Takes a DataFrame and returns a DataFrame.
    
    Parameters
    ----------
    df: A Pandas DataFrame, e.g. pd.read_hdf('/data/database/sflights.h5', 'table')
    
    Returns
    -------
    A Pandas DataFrame. It's a pivot table of flight cancellations
    as a function of month and Day. See assignment 12.2 instructions.
    '''
    
    #### your code goes here ###
    temp = df[df["depart"] == "ORD"]
    temp['dDelay'] = (temp['dDelay'] >= 15).astype(int) 
    temp = temp.groupby(["month", "Day"]).sum()
    temp.reset_index(inplace=True)  # reset the index
    result = temp.pivot("month", "Day", "dDelay")
    # my code ends here
    
    return result
````

````python
pivoted = pivot_table(data)
print(pivoted)
````
