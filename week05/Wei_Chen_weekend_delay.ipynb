```python
%matplotlib inline
import numpy as np, matplotlib.pyplot as plt
import csv

filename = '/data/airline/2001.csv'

with open(filename, encoding = 'latin-1') as my_file:
    air_csv = csv.reader(my_file, delimiter = ',') # use csv module to read this file
    heading = next(air_csv) # get heading of this form
    index_of_DayOfWeek = heading.index("DayOfWeek") 
    index_of_DepDelay = heading.index("DepDelay") 
    
    weekend_delay_info = { '5':[], '6':[], '7':[] } # initialization
    Fri_delay = Sat_delay = Sun_delay = [] # initialization
    
    for row in air_csv:
        if row[index_of_DayOfWeek] != "NA" and row[index_of_DepDelay] != "NA" and 5 <= int(row[index_of_DayOfWeek]) <= 7:
            weekend_delay_info[row[index_of_DayOfWeek]].append(int(row[index_of_DepDelay]))

Fri_delay = weekend_delay_info['5'] # aliasing
Sat_delay = weekend_delay_info['6'] # aliasing
Sun_delay = weekend_delay_info['7'] # aliasing

fig, ax = plt.subplots()
fig.set_size_inches((15,10))
ax.hist((Fri_delay, Sat_delay, Sun_delay), bins = 10, log = True, label=('Friday', 'Saturday', 'Sunday'), \
        color=('Red', 'Blue', 'Green'))
min_delay = min(Fri_delay + Sat_delay + Sun_delay)
max_delay = max(Fri_delay + Sat_delay + Sun_delay)

text_size = 14
ax.set_title("The Number of DepartureDelay on Weekends by Wei Chen", size = text_size)
ax.set_xlabel("Departure Delay Time", size = text_size)
ax.set_ylabel("Number", size = text_size)
ax.set_xlim(min_delay - 100, max_delay + 100)
ax.grid(True)
ax.legend()
plt.show()
```
