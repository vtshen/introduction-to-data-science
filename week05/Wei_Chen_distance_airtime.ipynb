```python
%matplotlib inline
import numpy as np, matplotlib.pyplot as plt
import csv

filename = '/data/airline/2001.csv'
interval = 1000 # plot every 1000th point
with open(filename, encoding = 'latin-1') as my_file:
    air_csv = csv.reader(my_file, delimiter = ',') # use csv module to read this file
    heading = next(air_csv) # get heading of this form
    index_of_AirTime = heading.index("AirTime") 
    index_of_Distance = heading.index("Distance") # get the index of AirTime and Distance
    
    data_list = []
    count = 0
    for row in air_csv:
        if row[index_of_AirTime] != 'NA' and row[index_of_Distance] != 'NA':
            count += 1
            if count == interval:
                data_list.append([row[index_of_AirTime], row[index_of_Distance]])
                count = 0 # reset count every time we pick a data point
        
fig, ax = plt.subplots()
ax.plot(data_list[::interval][0], data_list[::interval][1])
ax.set_xlabel("AirTime")
ax.set_ylabel("Distance")
ax.set_title("Distance vs. AirTime by Wei Chen")
plt.show()
```
