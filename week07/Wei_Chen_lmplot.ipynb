```python
%matplotlib inline
import seaborn as sns, matplotlib.pyplot as plt, pandas as pd, numpy as np

filename = '/data/airline/temp2.csv'

df = pd.read_csv(filename, encoding = "latin1", usecols = ["DepDelay", "ArrDelay", "DayOfWeek"])

```

```python
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True)
xlimit = ylimit = (-100, 200)
g1 = sns.regplot("DepDelay", "ArrDelay", df, ax=ax1)
g1.set_xlim(xlimit)
g1.set_ylim(ylimit)
g1.set_xlabel("Departure Delay (min)")
g1.set_title("The scatter plot of ArrDelay vs. DepDelay")
g2 = sns.boxplot(df["ArrDelay"], df["DayOfWeek"], ax=ax2)
g2.set_ylim(ylimit)
g2.set_title("The box plot of daily arrival delays")

f.set_size_inches(15,8)
f.tight_layout()
```
